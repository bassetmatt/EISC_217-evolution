# Plots tutorial

## Basic plotting

```julia
using Plots
x = 1:10; y = rand(10); # These are the plotting data
plot(x, y)
```

It's possible to have two series of data like this :

```julia
x = 1:10; y = rand(10, 2) # 2 columns means two lines
plot(x, y)
```

Plots are mutable. You can use `plot!(x,y)`<br>
It's possible to add another series to the plot like this :

```julia
z = rand(10)
plot!(x, z)
```

We can also use an explicit variable `p` for the plot

```julia
x = 1:10; y = rand(10, 2) # 2 columns means two lines
p = plot(x, y)
z = rand(10)
plot!(p, x, z)
```

## Plot attributes
Plots have a lot of possibilities, for example `plot(x,y,z)` will create a 3D plot and `plot(x,y,attribute=value)` will create a 2D plot with some attributes.<br>
Amongus those possibilities you can add Tiles, labels and legends

```julia
x = 1:10; y = rand(10, 2) # 2 columns means two lines
plot(x, y, title = "Upper Text", label = ["Line 1" "Line 2"], lw = 3)
```

Labels are also mutable using the `!` notation

```julia
xlabel!("My x label")
```

It implicitly uses `Plots.CURRENT_PLOT` and modifies `attribute!(p,value)`

## Plotting backends
Actually `Plots` is a meta package because it uses a lot of different librairies, the plotting librairy behind is called the backend.

We can use differnt backends

```julia
x = 1:10; y = rand(10, 2) # 2 columns means two lines
plotlyjs() # Set the backend to Plotly
# This plots into the web browser via Plotly
plot(x, y, title = "This is Plotted using Plotly")
```

There's also the GR backend

```julia
gr() # Set the backend to GR
# This plots using GR
plot(x, y, title = "This is Plotted using GR")
```

## Changing the plotting series
Okay lines are fine now, let's try making scatter points plots, from now we will add the GR backend. To use scatter points series uses the `seriestype` attribute. 

```julia
gr() # We will continue onward using the GR backend
plot(x, y, seriestype = :scatter, title = "My Scatter Plot")
```

For each type of series, there's a built-in function hat automatically generates the plot.

```julia
scatter(x,y,title = "My scatter Plot")
```

## Plotting in Scripts
Now that we can do plots, we can have a scrit to make them

```julia
display(plot(x, y))
```

## Combining multiple plots as subplots
We can have multiple plots together grouped as subplots<br>
The first method is to define a layout which will split series. The `layout` command takes in a 2-tuple `layout=(N,M)` which builds an N*M grid of plots, for example `layout=(4,1)` will get four rows of plots, each with one series.

```julia
y = rand(10, 4)
plot(x, y, layout = (4, 1))
```

We can also make a plot with subplots being different objects stored in explicit variables

```julia
p1 = plot(x, y) # Make a line plot
p2 = scatter(x, y) # Make a scatter plot
p3 = plot(x, y, xlabel = "This one is labelled", lw = 3, title = "Subtitle")
p4 = histogram(x, y) # Four histograms each with 10 points? Why not!
plot(p1, p2, p3, p4, layout = (2, 2), legend = false)
```

Now we are pretty much done with the basics, one last point will be about recipes, they're extensions to the `Plots.jl`, they add :
1. New `plot` commands via **user recipes**
2. Default interpretation of julia types via **types recipes** 
3. New functions for generating plots via **plot reccipes**
4. New series types via **series recipes**
